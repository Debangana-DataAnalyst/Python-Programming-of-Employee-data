```python
#open the json file
json_file="employee.json"
with open(json_file,"r") as j:
    j_file=json.load(j)
j_file
```

```python
#reading the json file
data=j_file['objects']
data=pd.DataFrame(data)
data
```

```python
#creating sqlite database data.db
conn=sqlite3.connect("data.db")
cursor=conn.cursor()
cursor
```

```python
#creating a table named employee_1
create_table=''' Create table IF NOT EXISTS employee_1(ID INT, JobTitle VARCHAR(50), EmailAddress VARCHAR(50), FirstNameLastName VARCHAR(50),vaccinated BOOLEAN);'''
cursor.execute(create_table)
```

```python
#Populating the data received from json into a nested list
for obj in j_file['objects']:
    cursor.execute('''INSERT INTO employee_1(ID,JobTitle,EmailAddress,FirstNameLastName,vaccinated) VALUES (?,?,?,?,?)''',(obj['ID'], obj['JobTitle'],
                                                                                                                          obj['EmailAddress'],obj['FirstNameLastName'],obj['vaccinated']))
    conn.commit()
    
```

```python
drop_table="DROP TABLE IF EXISTS employee"
cursor.execute(drop_table)
conn.commit
```

```python
#renamed employee_1 to employee
old_table_name='employee_1'
new_table_name='employee'
rename_table=(f" ALTER TABLE {old_table_name} RENAME TO {new_table_name};")
cursor.execute(rename_table)
conn.commit
```

```python
df=pd.read_sql_query("SELECT * FROM employee",conn) #reading the sql data back into pandas
```
